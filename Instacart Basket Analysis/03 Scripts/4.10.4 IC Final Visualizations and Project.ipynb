#### Part 1.3.2

```python
# Creation of a frequency dataframe for a regional chart

region_spending_crosstab = pd.crosstab(insta_proj['spending_flag'], insta_proj['region'],dropna = False)
```

```python
# creating the value for a saveable bar chart

region_spending_bar = region_spending_crosstab.plot(kind = 'bar',figsize=(10,6),color=['#08737f', '#000000', '#089f8f']) 
region_spending_bar.legend(title = 'Region') 
plt.title('Customer spending average in each region', fontsize = 20) 
plt.ylabel('Average amount spent (in USD)',fontsize=14) 
plt.xlabel('Type of spender',fontsize=14) 
plt.xticks(rotation=0,fontsize=12)
```

```python
# exporting the new figure

region_spending_bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'reg_freq.png'))
```
