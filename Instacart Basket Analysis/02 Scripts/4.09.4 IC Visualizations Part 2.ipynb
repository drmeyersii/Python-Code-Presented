```python
# Import libraries

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Path creation

path = r'/Users/nekow/Documents/Instacart Basket Analysis'
```

```python
ords_prods_all = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_all.pkl'))
```

## 1. Complete the same charts as practiced

```python
# Creating a bar chart

bar = ords_prods_all['orders_day_of_week'].value_counts().sort_index().plot.bar(color = ['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
plt.title('Busiest ordering days', fontsize = 20) 
plt.ylabel('Saturday - Friday',fontsize=14) 
plt.xlabel('Saturday - Friday',fontsize=14) 
plt.xticks(rotation=0,fontsize=12)
```

```python
# Exporting the bar chart

bar.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_orders_dow.png'))
```

```python
# Creating a scatterplot for outliers

sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_all)
```

```python
# Viewing outliers in the dataframe

ords_prods_all.loc[ords_prods_all['prices'] > 100]
```

```python
# Marking the outliers as missing

ords_prods_all.loc[ords_prods_all['prices'] > 100, 'prices'] = np.nan
```

```python
ords_prods_all['prices'].max()
```

```python
# Creating a histogram with more bins

hist_2 = ords_prods_all['prices'].plot.hist(bins = 70)
plt.title('Frequency of Ordering', fontsize = 20) 
plt.ylabel('Frequency(in millions)',fontsize=14) 
plt.xlabel('Hour of order(Military Standard Time)',fontsize=14) 
plt.xticks(rotation=0,fontsize=12)
```

```python
# Creating a line plot graph

line = sns.lineplot(data = ords_prods_all, x = 'orders_day_of_week',y = 'prices')
plt.title('Average price of items on day of purchase', fontsize = 20) 
plt.ylabel('Average product price (in USD)',fontsize=14) 
plt.xlabel('Friday - Saturday',fontsize=14) 
plt.xticks(rotation=0,fontsize=12)
```

```python
# Exporting the chart

line.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'line_orders_dow.png'))
```

### Step 6 of Exercise 4.9

```python
# Lineplot for ages and dependants

ages_line = sns.lineplot(data = ords_prods_all, x = 'age',y = 'n_dependants')
plt.title('Age and Average number of dependants', fontsize = 20) 
plt.ylabel('Average number of dependants',fontsize=14) 
plt.xlabel('Customer Age',fontsize=14) 
plt.xticks(rotation=0,fontsize=12)
```

```python
# Exporting the chart

ages_line.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'line_ages.png'))
```

## The correlation between Age and number of dependants seems to be extremely eratic and doesn't show very clear distinctions between the 2 variables

### Step 7 of Exercise 4.9

```python
# Scatterplot for ages and income

ages_scatter = sns.scatterplot(x = 'age', y = 'income',data = ords_prods_all) 
plt.title('Average age and income', fontsize = 20) 
plt.ylabel('Yearly Income(in USD)',fontsize=14) 
plt.xlabel('Customer Age',fontsize=14) 
plt.xticks(rotation=0,fontsize=12)
```

```python
# Exporting the chart

ages_scatter.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'scatter_ages.png'))
```

## It's clear that older ages with higher income have more consistent orders, although the more consistent orders happen to be lower income households across all ages.
